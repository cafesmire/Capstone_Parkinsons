### Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Goal: Visualizations

### Downsampled

```python
df_down = pd.read_csv('downsample_new_feature.csv', index_col='Time')
```

```python
df_down.head()
```

```python
numeric_df = df_down
numeric_df.corr()
```

```python
sns.heatmap(numeric_df.corr(), annot=numeric_df.rank(axis='columns').iloc[:9])
```

```python
sns.catplot(data=numeric_df, x='Event', kind='count', col='Medication')
```

```python
sns.catplot(data=numeric_df, x='Event', kind='count', col='Sex')
```

```python
sns.relplot(numeric_df, x=numeric_df.index, y='AccAP', kind='line')
```
